```python
def pre_process(text):
    def lemmatize(doc):
        doc = re.sub(patterns, ' ', doc)
        tokens = []
        for token in doc.split():
            if token.isalpha() and token not in stop_words:
                token = token.strip()
                token = morph.normal_forms(token)[0]       
                tokens.append(token)
        return tokens
    text=text.lower()
    text=re.sub("&amp;lt;/?.*?&amp;gt;"," &amp;lt;&amp;gt; ",text)
    text=re.sub("(\\d|\\W)+"," ",text)
    # text = [word for word in text if len(word) >= 3]
    text = lemmatize(text)
    return ' '.join(text)
```

## Dataset classes

```python
class RamblerDataset(Dataset):
    def __init__(self, path, tokenizer=None):
        self.data = pd.read_csv(path, sep=";", names=["Title", "Description", "Article Text", "Keywords"])
        self.keywords = self.data.Keywords.apply(lambda x: list(map(lambda y: y[1:-1].lower(), x[1:-1].split(", "))))
        self.tokenizer = tokenizer
    def __len__(self):
        return self.data.shape[0]

    def tokenize(self, text):
        if self.tokenizer is None:
            return {"text" : text}
        return self.tokenizer(text, return_tensors='pt', padding='max_length', truncation=True, max_length=150)

    def __getitem__(self, index):
        row = self.data.iloc[index]
        united_text = "; ".join([row.Title, row.Description, row["Article Text"]])
        items = {k: v.reshape(-1) for k, v in self.tokenize(united_text).items()}
        keywords = get_keyword_vector(self.keywords[index])
        items["keywords"] = torch.tensor(keywords) if keywords else None
        return items
```

```python
class RussianNewsDataset(Dataset):
    def __init__(self, path, tokenizer=None, only_keywords=True, only_title=False):
        self.data = pd.read_csv(path)
        # source	title	text	publication_date	rubric	subrubric	tags
        self.only_keywords = only_keywords
        if only_keywords:
            self.data = self.data[self.data['tags'].notna()]
        if only_title:
            self.data = self.data[self.data['rubric'].notna()]
        self.keywords = self.data.tags.str.lower().str.split(", ")
        self.keywords = np.array([get_keyword_vector(self.keywords.iloc[index]) for index in range(len(self.keywords))])
        indices = [i for i, k in enumerate(self.keywords) if sum(k) > 1]
        self.data.reset_index(inplace=True)
        self.data = self.data.loc[indices, :]
        self.keywords = self.keywords[indices]        
        self.data.reset_index(inplace=True)
        self.tokenizer = tokenizer

    def __len__(self):
        return self.data.shape[0]

    def tokenize(self, text):
        if self.tokenizer is None:
            return {"text" : text}
        return self.tokenizer(text, return_tensors='pt', padding='max_length', truncation=True, max_length=150)

    def __getitem__(self, index):
        row = self.data.iloc[index]
        united_text = "; ".join([row.title, "", row.text])
        items = {k: v.reshape(-1) for k, v in self.tokenize(united_text).items()}
        #print(f"index: {index}")
        #print(f"keywords: {self.keywords[index]}")
        if self.only_keywords:
            # keywords = get_keyword_vector(self.keywords[index])
            items["keywords"] = torch.tensor(self.keywords[index])
            #items["keywords"] = torch.tensor(get_keyword_vector(self.keywords[idx]))
        return items
```

```python
class VkMailDataset(Dataset):
    def __init__(self, path, tokenizer=None):
        self.data = pd.read_excel(path, sheet_name = list(range(4)))
        self.keywords = self.data.Keywords.str.lower().str.split(", ")
        self.tokenizer = tokenizer

    def __len__(self):
        return self.data.shape[0]

    def tokenize(self, text):
        if self.tokenizer is None:
            return {"text" : text}
        return self.tokenizer(text, return_tensors='pt', padding='max_length', truncation=True, max_length=150)

    def __getitem__(self, index):
        row = self.data.iloc[index]
        united_text = "; ".join([row.Title, row.Description, row["Article Text"]])
        items = {k: v.reshape(-1) for k, v in self.tokenize(united_text).items()}
        items["keywords"] = torch.tensor(get_keyword_vector(self.keywords[index]))
        return items
```

```python
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output['last_hidden_state']
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    return sum_embeddings / sum_mask
```
